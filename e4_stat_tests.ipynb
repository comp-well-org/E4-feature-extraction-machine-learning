```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_rel

# read data from saved features files
base_merged = pd.read_csv(r'./featuresData/base_merged.csv').set_index(["participant"], drop=True)
session_merged = pd.read_csv(r'./featuresData/session_merged.csv').set_index(["participant"], drop=True)
base_merged
test_result = {}
base_merged
```

T-test on acc magnitude

```python
base_acc = base_merged["AccelMagnitude"].groupby(base_merged.index).mean()
session_acc = session_merged["AccelMagnitude"].groupby(session_merged.index).mean()
stat, p = ttest_rel(base_acc, session_acc)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
test_result["AccelMagnitude"] = [p]
```

T-test on temperature

```python
base_temp = base_merged["Temp"].groupby(base_merged.index).mean()
session_temp = session_merged["Temp"].groupby(session_merged.index).mean()
stat, p = ttest_rel(base_temp, session_temp)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
test_result["Temp"]=[p]
```

T-test on IBI

```python
# baseline 1,37 session 26 are missing
base_IBI = pd.read_csv(r'./featuresData/single_base_IBI.csv').set_index(["participant"], drop=True)
session_IBI = pd.read_csv(r'./featuresData/single_session_IBI.csv').set_index(["participant"], drop=True)
base_IBI = base_IBI.groupby(base_IBI.index).mean().drop(26)
session_IBI = session_IBI.groupby(session_IBI.index).mean().drop([1,37])

for col in base_IBI:
    stat, p = ttest_rel(base_IBI[col], session_IBI[col])
    test_result[col]=[p]
    print('test on',col,': stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05:
        print('Probably the same distribution')
    else:
        print('Probably different distributions')
```

```python
result = pd.DataFrame(test_result)
result
```

```python
# eda peaks
base_eda = base_merged["peaks"].groupby(base_merged.index).mean()
session_eda = session_merged["peaks"].groupby(session_merged.index).mean()
stat, p = ttest_rel(base_eda, session_eda)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
test_result["peaks"]=[p]

# eda magnitude
base_eda = base_merged["filtered_eda"].groupby(base_merged.index).mean()
session_eda = session_merged["filtered_eda"].groupby(session_merged.index).mean()
stat, p = ttest_rel(base_eda, session_eda)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
test_result["eda"]=[p]
```
