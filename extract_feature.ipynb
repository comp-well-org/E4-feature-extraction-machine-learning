```python
import numpy as np
import pandas as pd
import datetime
import time
import os
import matplotlib.pyplot as plt
import re

from extract_E4_feature import extract_E4_features
```

```python
# a dictionary that maps participation number to string of baseline dataReader objects
baseline={}
# a dictionary that maps participation number to string of session session dataReader objects
session={}
# a dictionary that maps participation number to
baseline_roots={}
session_roots={}

# parse participant numbers and fill in the above two maps
for root, dirs, files in os.walk("./P02012019 - Sensory Stimulation (Frequency-dependent)/Participants Empatica Data_/"):
    i = root.find("part")+4
    if root[i+1] == '_':
        i = int(root[i])
    else:
        i = int(root[i:i+2])
    if "baseline" in root:
#         baseline[i] = dr(root+'/')
        baseline_roots[i] = root
    elif "session" in root and "fake" not in root:
#         session[i] = dr(root+'/')
        session_roots[i] = root
```

Extract EDA, ACC, and TEMP features

```python
%%time
thresh,offset,start_WT,end_WT = .02,1,4,4

settings_dict = {'threshold': thresh,
                'offset': offset,
                'rise time': start_WT,
                'decay time': end_WT}

# Extracting ACC features
baseList=[]
# basePeak=[]
baseIBI = []
ssnList=[]
ssnIBI=[]
for i in range(1,max(baseline_roots.keys())+1):
# for i in range(1,2):
    # processing baseline
    base_features_pd, base_peak_pd, base_ibi = extract_E4_features(baseline_roots[i], i, settings_dict)
    baseList.append(base_features_pd)
#     basePeak.append(base_peak_pd)
    baseIBI.append(base_ibi)
    
    #processing session
    ssn_features, ssn_peak, ssn_ibi = extract_E4_features(session_roots[i], i, settings_dict)
    ssnList.append(ssn_features)
#     basePeak.append(base_peak_pd)
    ssnIBI.append(ssn_ibi)
print("finished extracting features")
```

```python
base_pd = pd.concat(baseList)
ssn_pd = pd.concat(ssnList)
# base_peak_pd = pd.concat(basePeak)
# ssn_peak_pd = pd.concat(ssnPeak)
IBI_base_pd = pd.concat(baseIBI)
IBI_ssn_pd = pd.concat(ssnIBI)

base_merged = base_pd.merge(IBI_base_pd, on=['participant','Time'], how='outer')
ssn_merged = ssn_pd.merge(IBI_ssn_pd, on=['participant','Time'], how='outer')
```

```python
base_merged.to_csv (r'./features/base_merged.csv', index = False, header=True)
ssn_merged.to_csv (r'./features/session_merged.csv', index = False, header=True)
```
